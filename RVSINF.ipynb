```python
import pandas as pd
import matplotlib.pyplot as plt
file='datastroke.csv'
df=pd.read_csv(file)
req=['DRSISC','DRSH','DRSUNK','RVISINF']
data=df[req]
data.dropna().head()
```

```python
target = data[['DRSISC', 'DRSH','DRSUNK','RVISINF']]
target.dropna()
strk1 = []
for row in target[['DRSISC', 'DRSH','DRSUNK']].values:
    row = list(row)
    try:
        strk1.append(row.index('Y'))
    except:
        strk1.append(-1)
strk1
```

```python
strk2= []
for row in target[['RVISINF']].values:
    row = list(row)
    try:
        strk2.append(row.index('Y'))
    except:
        strk2.append(-1)
strk2[1:5]
```

```python
transactions = pd.DataFrame({
    'Recurrent_Stk': strk1,
    'Dead_Tissue': strk2
})
```

```python
import matplotlib.pyplot as plt
result = transactions.groupby('Recurrent_Stk')['Dead_Tissue'].value_counts()/transactions.groupby('Recurrent_Stk')['Dead_Tissue'].count() * 100
result
```
