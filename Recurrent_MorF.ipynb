```python
import pandas as pd
import matplotlib.pyplot as plt
file='datastroke.csv'
df=pd.read_csv(file)
req=['SEX','DRSISC']
data=df[req]

data=data[(data['DRSISC']=='Y')]

rec_ISC= data.groupby('SEX')['DRSISC'].value_counts()
rec_ISC.plot.bar()
plt.show()
```

```python
req=['SEX','DRSH']
data=df[req]
data=data[(data['DRSH']=='Y')]

rec_HAE= data.groupby('SEX')['DRSH'].value_counts()
rec_HAE.plot.bar()
plt.show()
```

```python
req=['SEX','DRSUNK']
data=df[req]

data=data[(data['DRSUNK']=='Y')]

rec_ISC= data.groupby('SEX')['DRSUNK'].value_counts()
rec_ISC.plot.bar()
plt.show()
```

```python
req=['SEX','DRSISC','DRSH','DRSUNK']
data=df[req]
rec_ISC= data.groupby('SEX')['DRSUNK'].value_counts()
rec_ISC.plot.bar()
plt.show()
```
